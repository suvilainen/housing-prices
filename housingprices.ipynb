```
from google.colab import files
uploaded = files.upload()
```

```
import pandas as pd

import io
size = 81
train = pd.read_csv(io.BytesIO(uploaded['train (1).csv'])).fillna(0)
print("train size " + str(train.shape))
trainSize = train.shape[0]
test = pd.read_csv(io.BytesIO(uploaded['test (1).csv'])).fillna(0)
print("test size " + str(test.shape))
testSize = test.shape[0]
data = pd.concat([train[train.columns[0:size-1]], test])
```

```
data = pd.get_dummies(data)
trainX = data.iloc[0:(trainSize), 0:(size-1)]

trainY = train[train.columns[size-1:size]]
testX = data.iloc[(trainSize):(testSize+trainSize), 0:(size-1)]
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

model = LinearRegression()
scores = []
kfold = KFold(n_splits=3, shuffle=True, random_state=42)
for i, (train, test) in enumerate(kfold.split(trainX, trainY)):
 model.fit(trainX.iloc[train,:], trainY.iloc[train,:])
 score = model.score(trainX.iloc[test,:], trainY.iloc[test,:])
 scores.append(score)
print(scores)
```

```
import numpy as np
model.fit(trainX, trainY)
testY = model.predict(testX)
print(testY)
moi = np.array(testY.flatten())


output = pd.DataFrame({'Id': test.Id, 'SalePrice': moi})
output.to_csv('my_housing.csv', index=False)
print("Your submission was successfully saved!")
```
